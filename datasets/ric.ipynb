```python
import os

def count_subfolders_files(folder_path):
    if not os.path.exists(folder_path) or not os.path.isdir(folder_path):
        print("Invalid folder path!")
        return

    for root, dirs, files in os.walk(folder_path):
        folder_name = os.path.relpath(root, folder_path)

        # Counting subfolders and files inside the current folder
        subfolder_count = len(dirs)
        file_count = len(files)

        print(f"Folder Name: {folder_name}")
        print(f"Number of Subfolders: {subfolder_count}")
        print(f"Number of Files: {file_count}")
        print("-" * 30)

# Replace 'path/to/your/folder' with the path of the folder you want to analyze
folder_path = 'train'
count_subfolders_files(folder_path)
```
